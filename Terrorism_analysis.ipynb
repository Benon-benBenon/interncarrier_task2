```python
# Assuming df is your DataFrame containing the data   
            
         
selected_columns =['eventid', 'iyear', 'imonth', 'iday', 'extended', 'country', 'region',
       'vicinity', 'crit1', 'crit2', 'crit3', 'success', 'suicide',
       'attacktype1', 'targtype1', 'individual', 'weaptype1', 'property',
       'INT_LOG', 'INT_IDEO', 'INT_MISC', 'INT_ANY']  # Replace column names accordingly

# Select only the desired columns
data_to_plot = final_df[selected_columns]

# Create boxplot using seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_to_plot,orient='h')
plt.title('Boxplot of Selected Columns')
plt.xlabel('Columns')
plt.ylabel('Values')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
def outliers_remove(final_df, val):
    q1 = final_df[val].quantile(0.25)
    q3 = final_df[val].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
    
    # Identify outliers using the bounds
    new_df = final_df.loc[(final_df[val] > lower_bound) & (final_df[val] < upper_bound)]
    print(new_df)
    
    # Return indices of outliers
    return new_df
for col_value in selected_columns:
    res_df= outliers_remove(final_df, col_value)
```
